# Handling missing values

```python
df.isna().sum()
```

```python
df.shape
```

```python
#dropping comments column so axis is 1
df_imp=df.drop(['Comments'], axis=1)
df_imp.shape

# #removing from categorical list as well
catcol.remove("Comments")
```

```python
#removing the rows where more than 5 values missing
df[df.isna().sum(axis=1)>5]
```

```python
df_imp=df_imp.drop([3,339],axis=0)
df_imp.shape
```

```python
#till here we dropped the 2 rows and a column
#but still there is missing data present
df.isna().sum()
```

```python
#impot library that is used in data science for imputation,preprocessing
#median is more preferred as the data might be left or right skewed 
from sklearn.impute import SimpleImputer

imp_median=SimpleImputer(missing_values=np.nan,strategy='median')
imp_mode=SimpleImputer(missing_values=np.nan,strategy='most_frequent')
```

```python
#creating dataframes 
df_num=pd.DataFrame(imp_median.fit_transform(df_imp[numcol]),columns=numcol)
df_cat=pd.DataFrame(imp_mode.fit_transform(df_imp[catcol]),columns=catcol)
```

```python
#now no missing values
df_num.isna().sum()
```

```python
df_cat.isna().sum()
```

```python
df_num.head()
```

```python
df_cat.head()
```

```python
#creating the dataframe back
df_imp=pd.concat([df_num,df_cat],axis=1)
```

# Outlier Analysis

```python
df.boxplot();
```

```python
for i in numcol:
    plt.figure()
    sns.boxplot(x=df[i])
```

```python
Q1=df.quantile(0.25)
Q3=df.quantile(0.75)
IQR=Q3-Q1
print(((df<(Q1-1.5*IQR)) | (df>(Q3+1.5*IQR))).sum())
```

```python
outnum=[]
for i in numcol:
    q1=df[i].quantile(0.25)
    q3=df[i].quantile(0.75)
    iqr=q3-q1
    
    upper_limit=q3+1.5*iqr
    lower_limit=q1-1.5*iqr
    outnum.append([i,q1,q3,iqr,lower_limit,upper_limit])
    
    df[i]=np.where(df[i]>upper_limit,upper_limit,
                  np.where(df[i]<lower_limit,lower_limit,
                          df[i]))
```

```python
pd.DataFrame(outnum,columns=["variableName","Q1","Q3","IQR","lower_limit","upper_limit"])
```

# Encoding Variables

```python
df_imp[catcol]
```

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
# while labelencoder creates a sigle column
# if we have 11 unique values or variables present then it creates 10 unique columns for rest-onehotencoder
```

```python
#we need to decide which should go under onehot and label encoder
#since individual has 190 unique values it will create 189 new columns
#in date egg has 50 unique values and it will create 49 new columns
#we dont want that many new column for analysis
#so choose less unique values to be onehotencoder and more unique values for labelencoder
for i in catcol:
    print(i,df[i].nunique())
```

```python
#how we do labelencoding
le=LabelEncoder()
for i in ["Individual ID","Date Egg"]:
    df_imp[i]=le.fit_transform(df_imp[i])
```

```python
#we see the individual id and date egg with variables is now given numerical values
df_imp[catcol]
```

```python
for i in["studyName","Species","Region","Island","Stage","Clutch Completion","Sex"]:
    y=pd.get_dummies(df_imp[i], prefix=i)
    df_imp=pd.concat([df_imp,y], axis=1)
```
