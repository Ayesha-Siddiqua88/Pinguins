```python
import os
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("penguins_info.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
# multivariate shows relation between various variables
#with help of pairplot we can determine the relation btw all variables(having numerical values)
sns.pairplot(df)
plt.plot()
plt.show()
```

```python
# from above we notice the relation btw flipper length and body mass
# we see that the body mass increases with inc in flipper length
# we have univariate analysis i.e. for each variable histogram is plotted
# also bivariate analysis btw two varibles i.e. dots
```

```python
# correlation- if you change one unit of a variable by how much unit the other variable changes

plt.figure(figsize=(16,16))
sns.heatmap(df.corr(),annot=True)
plt.show()
```

```python
# from above plot if we change the sample number by 1 unit then the
#culmen length changes by -0.24 unit
```
